# Data Cleaning

```python
#Check duplicate rows
df.duplicated().sum()
```

```python
#count rows
df["ID"].count()
```

```python
#Drop duplicate rows
df = df.drop_duplicates()
df["ID"].count()
```

```python
#Delete rows before 1980
df= df[df['Prod. year'] >= 1990]
```

```python
#Rename Column name
df.rename(columns={'Prod. year': 'Production Year'}, inplace=True)
```

```python
#Column Split
df["Mileage (KM)"] = df["Mileage"].apply(lambda x: int(x.split(" ")[0]))
```

```python
#Drop Column
df.drop('Mileage', axis=1, inplace=True)
```

```python
# Replace value
df['Levy'] = df['Levy'].replace('-', 0) #Single Column
df.replace('-', 0, inplace=True) # Full DataFrame
```

```python
# Casting string to int
df["Levy"] = pd.to_numeric(df["Levy"])
```

```python
# create a new column that can calculate car age
df['Car Age'] = 2025 - df['Production Year']
```

```python
#To remove rows with very low frequency (i.e., rows that appear very few times) 
df = df[df['Fuel type'] != 'Hydrogen'] # 1 value
df = df[df['Category'] != 'Limousine'] # 11 values
```

```python
# Engine volume
# Remove 'Turbo' and strip any extra spaces 
# Everything only for casting to float
df['Engine volume'] = df['Engine volume'].str.replace(' Turbo', '', regex=False).str.strip()
df['Engine volume'] = df['Engine volume'].astype(float)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import MinMaxScaler
```

```python
X = df.drop(["ID", "Price", "Doors", "Production Year", "Color", "Category"], axis=1)
Y = df["Price"]
```

```python
X.info()
```

```python
transformer = ColumnTransformer(transformers=[
    ('t1', OneHotEncoder(sparse_output=False, drop='first'), 
             ['Model', 'Manufacturer', 'Leather interior', 'Fuel type', 'Gear box type', 'Drive wheels', 'Wheel']),
    ('t2', MinMaxScaler(),
             ['Levy', 'Engine volume', 'Mileage (KM)', 'Cylinders', 'Airbags', 'Car Age'])
    ], remainder='passthrough')
```

```python
# Convert target variable to NumPy array
y = np.array(Y)
# Apply the transformation
x = transformer.fit_transform(X)
```

```python
######################################################################################################################
```

# Model Tranning

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

print(x.shape)
print(y.shape)

print(x_train.shape)
print(y_train.shape)

print(x_test.shape)
print(y_test.shape)
```

# <span style="color:red">Linear Regression</span>

```python
from sklearn.linear_model import LinearRegression
model1 = LinearRegression()
model1.fit(x_train, y_train)
```

```python
x_test[0]
```

```python
y_test
```

```python
y_pred1 = model1.predict(x_test)
y_pred1
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.metrics import mean_squared_error,r2_score

mae = mean_absolute_error(y_test, y_pred1)
r2 = r2_score(y_test, y_pred1)
```

```python
print("MAE: ", mae)
print("R2: ",r2)
```

# <span style="color:#90ee90">RandomForestRegressor</span>
<span style="color:blue">Best model trainning</span>

```python
from sklearn.ensemble import RandomForestRegressor

model2 = RandomForestRegressor(n_estimators=100, random_state=42)
model2.fit(x_train, y_train)
```

```python
#Actual Output
print(y_test)
#Predicted Output
y_pred2 = model2.predict(x_test)
print(y_pred2)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.metrics import mean_squared_error,r2_score

mae = mean_absolute_error(y_test, y_pred2)
r2 = r2_score(y_test, y_pred2)
print("MAE is: ",mae)
print("R2: ",r2)
```
